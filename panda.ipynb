## Tidy GDP

Manipulate the GDP.csv file and make it tidy, the result should be a pandas dataframe with the following columns:
* Country Name
* Country Code
* Year
* GDP

```python
df=pd.read_csv('https://raw.githubusercontent.com/gumdropsteve/intro_to_python/main/day_09/data/gdp.csv ')
df
```

```python
#show the colmn of the csv file
df.columns
```

```python
#Show the years column
df.columns[4:62]
```

```python
# Explore Years colomn

years  = [df.iloc[0  , 4:62] for i in range(1960,2017,1)]
   
years
```

```python
#remove unnessesery column
GDP_df = df.drop(columns=['Indicator Name', 'Indicator Code', 'Unnamed: 62'])
GDP_df
```

```python
# use melt function to create Year colomn
GDP_df = GDP_df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='GDP')
GDP_df
```
